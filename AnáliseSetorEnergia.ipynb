# Criando os parâmetros de limite

```python
potenciareal=turbina['ActivePower(kw/h)'].tolist() #Transformando a coluna ActivePower em uma lista
print(potenciareal[0:15])
```

```python
potenciateorica=turbina['TheoreticalCurve(kw/h)'].tolist() #Transformando a coluna Theoretical Curve em uma lista
print(potenciateorica[0:15])
```

# Criar uma lista de potência máxima (valor da turbina teorica + 5%) e potência minima ( valor da turbina teorica - 5%)




```python
#Usaremos o FOR percorrendo cada registro de potencia na lista que criamos e adicionando +5% e -5%
```

Exemplo: Em um registro de potencia teorica temos o valor de 1000. O programa percorrera cada registro e jogará na variável de pot_max o valor multiplicado por 1.05, que significa adicionar 5%. Ou seja, a potencia máxima que o valor de potencia teorica de 1000 pode atingir é de 1000*1.05 = 1050. O mesmo será feito com a potencia minima, que será 1000*0.95=950.

```python
pot_max= []
pot_min=[]
dentro_limite=[]


for potencia in potenciateorica:
  pot_max.append(potencia*1.05)
  pot_min.append(potencia*0.95)

#Para cada potencia de indice p na lista potencia real temos um SE, onde se a potencia real por maior ou igual a potencia minima correspondente ao mesmo indice p e a potencia for menor ou igual a potencia máxima de indice p, diremos que a potencia está dentro do limite
for p, potencia in enumerate(potenciareal):
    if potencia >= pot_min[p] and potencia<=pot_max[p]:
        dentro_limite.append("Dentro")
    elif potencia==0:
      dentro_limite.append("Zero")
    else:
      dentro_limite.append("Fora")


#Verificando a porcentagem de registros de potencia real que estão dentro do limite
new_var = print(dentro_limite.count("Dentro")/len(dentro_limite))

new_var


#Para verificar se as 3 listas possuem o mesmo tamanho, verificando assim se todos os registros foram percorridos igualmente
print(len(pot_max), len(pot_min), len(potenciateorica))
print(pot_max, pot_min)
```

# Adicionando a lista "dentro do limite" ao dataframe

```python
turbina['Within the bounds']=dentro_limite
display(turbina)
```

# Plotando o gráfico com os parâmetros de limite configurados


"Hue" é usado para atribuir cores diferentes às observações com base em uma terceira variável. Iremos colorir os pontos do gráfico com base em se eles estão "Within the bounds" (Dentro dos limites) ou não.


```python
cores = {'Dentro': 'blue', 'Fora':'red', 'Zero':'orange' } #Atribuindo um dicionário onde cada valor representará uma cor no gráfico.
sns.scatterplot(data=turbina, x="WindSpeed(m/s)", y='ActivePower(kw/h)', hue='Within the bounds', s=5, palette=cores)
```
